```python
import os
from pdf2image import convert_from_path
from PIL import Image

def convert_pdf_to_jpg(pdf_path, output_folder, Poppler_PATH, header_crop=100, footer_crop=100):
    """
    pdf_path: Path to your PDF file
    output_folder: Where to save JPGs
    header_crop: Number of pixels to cut from the top
    footer_crop: Number of pixels to cut from the bottom
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Convert PDF pages to PIL Image objects
    # Use 300 DPI for high quality
    pages = convert_from_path(
        pdf_path,
        300,
        poppler_path=Poppler_PATH
    )
    for i, page in enumerate(pages):
        width, height = page.size

        # Define the cropping box: (left, top, right, bottom)
        # We keep full width, but trim the top (header) and bottom (footer)
        left = 0
        top = header_crop
        right = width
        bottom = height - footer_crop

        # Ensure we don't crop more than the image height
        if top < bottom:
            cropped_page = page.crop((left, top, right, bottom))

            # Save the file
            output_filename = f"page_{i + 1}.jpg"
            save_path = os.path.join(output_folder, output_filename)
            cropped_page.save(save_path, 'JPEG')
            print(f"Saved: {save_path}")
        else:
            print(f"Skipping page {i+1}: Crop margins are too large.")

# Upload your PDF to Colab first!
POPPLER_PATH=r"E:\LLAMAIDX-ENV\poppler-25.12.0\Library\bin"
pdf_file = '114碩士班招生簡章.pdf'
output_dir = 'results'

# 需要略過的header/footer需要自行調整
convert_pdf_to_jpg(pdf_file, output_dir, Poppler_PATH=POPPLER_PATH, header_crop=150, footer_crop=150)
```
